```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Load Results

```python
df_BKM = pd.read_csv('./pred_res_BKM.csv')
df_BGMM = pd.read_csv('./pred_res_BGMM.csv')
```

```python
R2_max_BKM = df_BKM.drop(columns=['Cluster']).max(axis=1)
R2_max_BGMM = df_BGMM.drop(columns=['Cluster']).max(axis=1)
```

## Box Plot 

```python
plt.figure(figsize=(8,6))
plt.boxplot([R2_max_BKM, R2_max_BGMM], labels=['R2 from BKM', 'R2 from BGMM'])
plt.scatter([1]*33, R2_max_BKM, color='c', alpha=0.5, marker='o')
plt.scatter([2]*30, R2_max_BGMM, color='m', alpha=0.5, marker='o')
plt.ylabel('R2', fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
