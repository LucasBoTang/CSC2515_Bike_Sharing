```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
TRI_DATA_DIR = './trip_data/'
CLI_DATA_DIR = './climate_data/'
```

## Load Data 

```python
df_climate = pd.read_csv(CLI_DATA_DIR+'climate.csv')
df_station = pd.read_csv(TRI_DATA_DIR+'stations_check.csv')
```

## Bad Weather 

### Rain 

```python
df_climate[df_climate['Moderate Rain'] == 1].groupby(['Year', 'Month', 'Day']).count()['Date/Time']
```

```python
df_climate[(df_climate['Year'] == 2019) & (df_climate['Month'] == 4) & (df_climate['Day'] == 19)]
```

```python
rainyday = df_station[(df_station['Year'] == 2019) & (df_station['Month'] == 4) & (df_station['Day'] == 19)] \
           .groupby('Time').sum()
rainyday
```

```python
avg = df_station[(df_station['Year'] == 2019) & (df_station['Month'] == 4)] \
      .groupby(['Day','Time']).sum().groupby('Time').mean()
avg
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(24), rainyday['Checkin per Hour'], marker='o', ls='--', color='c', lw=5, label='Rainy Day')
plt.plot(range(24), avg['Checkin per Hour'], marker='o', color='m', lw=5, alpha=0.5, label='Month Avg')
plt.legend(fontsize=18)
plt.xlabel('Hour', fontsize=24)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(24), rainyday['Checkout per Hour'], marker='o', ls='--', color='c', lw=5, label='Rainy Day')
plt.plot(range(24), avg['Checkout per Hour'], marker='o', color='m', lw=5, alpha=0.5, label='Month Avg')
plt.legend(fontsize=18)
plt.xlabel('Hour', fontsize=24)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

## Temp 

```python
temp = df_climate.groupby('Month').mean()
temp
```

```python
check = df_station.groupby('Month').sum()
check
```

```python
fig = plt.figure(figsize=(5,4))
 
ax1 = fig.add_subplot()
ax1.plot(range(1,13), temp['Temp (°C)'], color='b', lw=3, alpha=0.5, label='Temp')
ax1.set_xlabel('Month', fontsize=16)
ax1.set_ylabel('Temp (°C)', fontsize=16)
ax1.set_xticks(temp.index)
 
ax2 = ax1.twinx()
ax2.plot(check.index, check['Checkin per Hour'], color='r', lw=3, alpha=0.5, label='Checkin')
ax2.set_ylabel('Total Checkin', fontsize=16)

fig.legend(loc=2, fontsize=12)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
